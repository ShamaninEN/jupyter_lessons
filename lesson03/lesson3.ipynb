```python
# Задание 4
DATASET_PATH = 'creditcard.csv'
df = pd.read_csv(DATASET_PATH)
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import train_test_split
```

```python
df['Class'].value_counts(normalize=True)
```

```python
pd.options.display.max_columns = 100
```

```python
df.info()
```

```python
df.head(10)
```

```python
X = df.drop('Class', axis=1)
X.head(10)
```

```python
y = df['Class']
y.head()
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size = 0.30, random_state=100, stratify=y)
```

```python
X_train, X_test, y_train, y_test
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
clf = GridSearchCV(estimator=RandomForestClassifier(random_state=100),param_grid=parameters,scoring='roc_auc',cv=3)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_predict_proba = clf.predict_proba(X_test)
y_predict_proba
```

```python
y_pred_proba = y_predict_proba[:,1]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```
